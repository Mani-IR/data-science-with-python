# تمرین چهارم - طبقه‌بندی با درخت تصمیم و Naive Bayes

```python

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report

df = pd.read_csv('airport-codes.csv')

cols = ['latitude_deg','longitude_deg','elevation_ft','type']
df = df[cols].dropna()

X = df[['latitude_deg','longitude_deg','elevation_ft']]

le = LabelEncoder()
y = le.fit_transform(df['type'])

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)
```

```python

tree = DecisionTreeClassifier(random_state=42)
tree.fit(X_train, y_train)

pred_tree = tree.predict(X_test)

acc_tree = accuracy_score(y_test, pred_tree)

print("Accuracy Decision Tree:", acc_tree)
```

```python

nb = GaussianNB()
nb.fit(X_train, y_train)

pred_nb = nb.predict(X_test)

acc_nb = accuracy_score(y_test, pred_nb)

print("Accuracy Naive Bayes:", acc_nb)
```

```python

print("\nمقایسه نهایی")
print("Decision Tree:", acc_tree)
print("Naive Bayes:", acc_nb)

if acc_tree > acc_nb:
    print("درخت تصمیم بهتر عمل کرده است.")
else:
    print("Naive Bayes بهتر عمل کرده است.")
```
